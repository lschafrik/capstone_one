```python
def transform_height(col):
    ht = col.split("' ")
    ft = float(ht[0])
    inch = float(ht[1])
    return(12*ft)+inch
```

```python
type(transform_height("5' 8"))
```

```python
rtr_pd['height'] = rtr_pd['height'].map(lambda x: transform_height(x))
```

```python
from datetime import datetime

def transform_date(col):
    return datetime.strptime(col, '%B %d, %Y')
```

```python
rtr_pd['review_date'] = rtr_pd['review_date'].map(lambda x: transform_date(x))
```

```python
# central limit theorem with age - take 50 samples
age_list = list(rtr_pd['age'])
```

```python
print("Mean of Age Population: ", stats.mean(age_list))
print("Variation of Age Population: ", stats.pvariance(age_list))
print("Standard Deviation of Age Population: ", stats.pstdev(age_list))
print("Size of Age Population: ", len(age_list))
```

```python
agequery_fit = spark.sql("SELECT DISTINCT(age), fit, COUNT(fit) FROM rtr_age GROUP BY age, fit ORDER BY age")
```

```python
agequery_fit.show()
```

```python
fit_pd = agequery_fit.toPandas()
```

```python
fig, ax_fit = plt.subplots(figsize=(15,9))

sm = fit_pd[fit_pd['fit']=='small'][['age', 'count(fit)']]
ft = fit_pd[fit_pd['fit']=='fit'][['age', 'count(fit)']]
lg = fit_pd[fit_pd['fit']=='large'][['age', 'count(fit)']]

ax_fit.plot(sm['age'], sm['count(fit)'], label='Small')
ax_fit.plot(ft['age'], ft['count(fit)'], label='Fit')
ax_fit.plot(lg['age'], lg['count(fit)'], label='Large')
ax_fit.set_xlim(10,100)
ax_fit.legend(fontsize=12)
ax_fit.set_xlabel('Age')
ax_fit.set_ylabel('Count')
ax_fit.set_title('Fit Status by Age', fontweight='bold')
```
