```python
rtr_data = pd.read_csv('rtr_data')
```

```python
rtr_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
rtr_data
```

```python
# avg data for fit, rented_for, body_type, category
```

```python
rtr_spark = spark.createDataFrame(rtr_data[['age', 'weight_lbs', 'height_inches', 'size', 'body_type', 'category', 'rating', 'fit', 'rented_for', 'review_date']])
rtr_spark.createOrReplaceTempView('rtr')
```

```python
rentfor_query = spark.sql("SELECT rented_for, avg(age), avg(weight_lbs), avg(height_inches), avg(size), avg(rating) FROM rtr GROUP BY rented_for ORDER BY avg(age)")
```

```python
rf_pd = rentfor_query.toPandas()
rentfor_query.show()
```

```python
cat_query = spark.sql("SELECT category, avg(age), avg(weight_lbs), avg(height_inches), avg(size), avg(rating) FROM rtr GROUP BY category ORDER BY avg(age)")
cat_pd = cat_query.toPandas()
cat_query.show(68)
```

```python
cat_queryct = spark.sql("SELECT category, COUNT(category) FROM rtr GROUP BY category ORDER BY COUNT(category)")
```

```python
cat_queryct.show(68)
```

```python
rf_queryct = spark.sql("SELECT rented_for, COUNT(rented_for) FROM rtr GROUP BY rented_for ORDER BY COUNT(rented_for)")
rf_queryct.show()
```

```python
bt_queryct = spark.sql("SELECT body_type, COUNT(body_type) FROM rtr GROUP BY body_type ORDER BY COUNT(body_type)")
bt_queryct.show()
```

```python
# most popular category per season per year
```

```python
rtr_data['review_date']
```

```python
rtr_data['review_date'][0][:4]
```

```python
rtr_data['review_year'] = rtr_data['review_date'].map(lambda x: x[:4])
```

```python
rtr_data['review_year']
```

```python
rtr_data['review_month'] = rtr_data['review_date'].map(lambda x: x[5:7])
```

```python
rtr_data['review_month']
```

```python
rtr_dates_rf = rtr_data.groupby(['review_year', 'rented_for']).count()
rtr_dates_rf
```

```python
for yr in np.unique(rtr_data['review_year']):
    rf_yr = rtr_dates_rf.loc[yr]['fit']
    print(f'most pop for {yr} is {rf_yr.idxmax()}')
```

```python
rtr_mt_rf = rtr_data.groupby(['review_month', 'rented_for']).count()
rtr_mt_rf
```

```python
for mt in np.unique(rtr_data['review_month']):
    rf_mt = rtr_mt_rf.loc[mt]['fit']
    print(f'most pop for {mt} is {rf_mt.idxmax()}')
```

```python
rtr_age_rfct = rtr_data.groupby(['age', 'rented_for']).count()
rtr_age_rfct
```

```python
popcat_by_age = []
for age in np.unique(rtr_data['age']):
    age_df = rtr_age_rfct.loc[age]['fit']
    print(f'most popular category for age {age} is {age_df.idxmax()} with {age_df.max()} rentals')
    popcat_by_age.append([age, age_df.idxmax(), age_df.max()])
```

```python
popcat_by_age
```
