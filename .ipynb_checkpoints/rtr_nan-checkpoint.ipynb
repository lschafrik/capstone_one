```python
nan_data = pd.read_csv('rtr_nan_data')
nan_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
nan_data.describe()
```

```python
nan_data['weight'].fillna(0, inplace=True)
```

```python
nan_data['weight'] = nan_data['weight'].map(lambda x: str(x).rstrip('lbs'))
nan_data['weight'] = nan_data['weight'].astype(int)
```

```python
nan_data['height'].fillna('0\' 0"', inplace=True)
nan_data['height'] = nan_data['height'].map(lambda x: str(x).rstrip('"'))
```

```python
def transform_height(col):
        ht = col.split("' ")
        ft = float(ht[0])
        inch = float(ht[1])
        return(12*ft)+inch
```

```python
nan_data['height'] = nan_data['height'].map(lambda x: transform_height(x))
```

```python
from datetime import datetime

def transform_date(col):
    return datetime.strptime(col, '%B %d, %Y')
```

```python
nan_data['review_date'] = nan_data['review_date'].map(lambda x: transform_date(x))
```

```python
nan_data['age'].fillna(0, inplace=True)
```

```python
nan_data.rename(columns = {'height':'height_inches', 'weight':'weight_lbs', 'body type':'body_type', 'rented for':'rented_for', 'bust size':'bust_size'}, inplace=True)
```

```python
nan_data.isnull().sum()
```

```python
nan_data['bust_size'].fillna('Unknown', inplace=True)
nan_data['rating'].fillna(0.1, inplace=True)
nan_data['rented_for'].fillna('Unknown', inplace=True)
nan_data['review_text'].fillna('No text', inplace=True)
nan_data['body_type'].fillna('Unknown', inplace=True)
nan_data['review_summary'].fillna('No text', inplace=True)
```
